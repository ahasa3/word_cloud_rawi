# Preproccessing matan

## menghapus awalan matan yang tidak diperlukan

```python
def remove_first_word(split_word,word):
    for i in split_word:
        if i == word:
            split_word.remove(word)
            break
    return split_word
def remove_space(split_w):
    for i, w in enumerate(split_w):
        if w == " ":
            split_w.remove(i)
    return split_w
```

```python
remove_word = ["rasulullah","nabi",'rasul','bersabda',"shallallahu", "'alaihi","wasallam"]
new_matan = []
for i, matan in enumerate(new_df.matan.tolist()):
    splt = matan.split(' ')
    for i in remove_word:
        splt = remove_first_word(splt,i)
    new_matan.append(' '.join(splt).replace("'","").replace('-',' ').replace('   ',' ').replace('  ',' '))
```

```python
new_matan
```

## menimpa matan sebelumnya dengan matan baru yang sudah di preprocessing

```python
df_fix_matan = pd.concat([new_df.drop('matan',axis=1),pd.DataFrame({'matan':new_matan})], axis=1)
df_fix_matan
```

## Menemukan Matan yang diriwayatkan oleh perawi yang paling banyak muncul

```python
def find_matan_with_rawi(matan_, perawi):
    index = []
    for i, rawi in enumerate(matan_):
        if perawi in rawi:
            index.append(i)
    return index
```

```python
choosed_rawi = ['abu hurairah','ibnu abbas','ibnu syihab']
index_rawi = []
for i in choosed_rawi:
    idx = find_matan_with_rawi(df_fix_matan.matan.tolist(), i)
    index_rawi.append(idx)
abu_hurairah = df_fix_matan[df_fix_matan.id.isin(index_rawi[0])]
ibnu_abbas = df_fix_matan[df_fix_matan.id.isin(index_rawi[1])]
ibnu_syihab = df_fix_matan[df_fix_matan.id.isin(index_rawi[2])]
```

# Topik Modelling

```python
from sklearn.decomposition import LatentDirichletAllocation as LDA
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def print_Topics(model, feature_name, Top_Topics, n_top_words):
    for topic_idx, topic in enumerate(model.components_[:Top_Topics]):
        print("Topic #%d:" %(topic_idx+1))
        print(" ".join([feature_name[i]
                       for i in topic.argsort()[:-n_top_words - 1: -1]]))
```

```python
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()
```

## Abu Hurairah

```python
abu_hurairah['rmvstop'] = abu_hurairah['matan'].apply(lambda text: stopword.remove(text))
```

```python
abu_hurairah.rmvstop.head()
```

```python
PrcsLDA = abu_hurairah['rmvstop']
tf_vectorizer = CountVectorizer(max_df = 1.0, min_df = 1)
tf = tf_vectorizer.fit_transform(PrcsLDA)
tf_terms = tf_vectorizer.get_feature_names()
```

```python
n_topics = 5
lda = LDA(n_components = n_topics, learning_method = 'batch', random_state = 0).fit(tf)
```

```python
Top_Words = 10
print('Printing top {0} Topics, with top {1} Words:'.format(n_topics, Top_Words))
print_Topics(lda, tf_terms, n_topics, Top_Words)
```
